```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.normal(20, 1.5, 2000)
plt.hist(data,
        bins=20)
```

```python
data = np.random.random(2000)
plt.scatter(data, [x for x in range(1, 2001)])
```

```python
for i in range(4):
    x = np.linspace(0, 2 * np.pi, 400)
    y = np.sin(x ** 2)
    plt.subplot(2, 2, i+1)
    plt.tight_layout()
    plt.title(f'{i+1} subplot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x, y)
```
